# Task 1: Prediction Using Supervised ML

### Read Data

```python
df_task1 = pd.read_csv("../input/student-scores/student_scores - student_scores.csv")
df_task1.head()
```

```python
X = df_task1.drop(["Hours"],axis = 1)
y = df_task1["Hours"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
type(X_test)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
print("Score = ",str(reg.predict(pd.DataFrame([9.25]))[0]))
```

```python
from sklearn import metrics  
y_pred = reg.predict(X_test)
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

# Task 2 : Prediction Using Unsupervised ML 

```python
df_task2 = pd.read_csv("../input/irisdata/Iris.csv",index_col = "Id")
df_task2.head()
```

```python
df_task2.nunique()
```

```python
distortions = []
K = range (1,11)
X = df_task2.drop("Species",axis = 1)
y = df_task2["Species"]
for k in K:
    KmeansModel = KMeans(n_clusters=k)
    KmeansModel.fit(X)
    distortions.append(KmeansModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeanModel = KMeans(n_clusters=3)
kmeanModel.fit(X)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df_task2['Species']= label_encoder.fit_transform(df_task2['Species'])
```

```python
df_task2["Species"]
```

```python
df_task2['Y_Predict']=kmeanModel.predict(X)
#df_task2['target']=iris['target']
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(df_task2["SepalLengthCm"], df_task2["SepalWidthCm"], c=df_task2['Species'])
axes[1].scatter(df_task2["SepalLengthCm"], df_task2["SepalWidthCm"], c=df_task2['Y_Predict'], cmap=plt.cm.Set1)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('K_Means', fontsize=18)
```
